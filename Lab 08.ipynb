```python
data_3b = pd.read_csv("data/Lab08-3b-1.csv")
data_3b_2pole = pd.read_csv("data/Lab08-3b-2.csv")
data_3d = pd.read_csv("data/Lab08-3d.csv")
```

```python
plt.plot(data_3b["Frequency (Hz)"], np.power(10, data_3b["Channel 2 Magnitude (dB)"]/20), label="1 pole filter")
plt.plot(data_3b_2pole["Frequency (Hz)"], np.power(10, data_3b_2pole["Channel 2 Magnitude (dB)"]/20), label="2 pole filter")
plt.plot(data_3d["Frequency (Hz)"], np.power(10, data_3d["Channel 2 Magnitude (dB)"]/20), label="Chebyshev Sallen-Key")
plt.legend()
plt.xscale('log')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (X)")
```
